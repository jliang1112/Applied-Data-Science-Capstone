Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
# Create 'marker_color' column based on 'class' values
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
import folium
site_map.save("map.html")

# Add marker_cluster to the current site_map
site_map.add_child(marker_cluster)

# Loop through each launch record and add a Marker to the cluster
for index, record in spacex_df.iterrows():
    coordinate = [record['Lat'], record['Long']]
    label = f"{record['Launch Site']} - {'Success' if record['class'] == 1 else 'Failure'}"
    
    # Create and add marker
    marker = folium.Marker(
        location=coordinate,
        popup=label,
        icon=folium.Icon(color='white', icon_color=record['marker_color'], icon='info-sign')
    )
    
    marker_cluster.add_child(marker)

# Show the map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of Earth in kilometers
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.features import DivIcon
from folium.plugins import MousePosition
from math import radians, sin, cos, sqrt, atan2

# Distance calculation function (Haversine formula)
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # Radius of Earth in km

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Coordinates
launch_site_lat = 28.563197
launch_site_lon = -80.576820
destination_lat = 28.56203
destination_lon = -80.56766

# Calculate distance
distance = calculate_distance(launch_site_lat, launch_site_lon, destination_lat, destination_lon)

# Initialize map
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=15)

# Add MousePosition to display coordinates on hover
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' | Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=10,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)

# Add launch site marker
folium.Marker(
    [launch_site_lat, launch_site_lon],
    popup='Launch Site: CCAFS SLC-40',
    icon=folium.Icon(color='green')
).add_to(site_map)

# Add destination marker
folium.Marker(
    [destination_lat, destination_lon],
    popup='Coastline',
    icon=folium.Icon(color='blue')
).add_to(site_map)

# Draw PolyLine
folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [destination_lat, destination_lon]],
    color='blue',
    weight=2,
    dash_array='5,10'
).add_to(site_map)

# Add distance label
folium.Marker(
    [ (launch_site_lat + destination_lat)/2, (launch_site_lon + destination_lon)/2 ],
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html='<div style="font-size: 12px; color:#d35400;"><b>{:.2f} KM</b></div>'.format(distance),
    )
).add_to(site_map)

# Display map
from IPython.display import display
display(site_map)
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
```

```python
cities = [
    {"label": "Orlando", "lat": 28.5383, "lon": -81.3792},
    {"label": "Titusville", "lat": 28.6122, "lon": -80.8076},
]

railways = [
    {"label": "Rail Line A", "lat": 28.57205, "lon": -80.58528},
    {"label": "Rail Line B", "lat": 28.5900, "lon": -80.6000},
]

highways = [
    {"label": "Highway 1", "lat": 28.6055, "lon": -80.8077},
    {"label": "Interstate 95", "lat": 28.6035, "lon": -80.8130},
]


def find_closest(infra_list, launch_lat, launch_lon):
    min_distance = float('inf')
    closest = None
    for point in infra_list:
        dist = calculate_distance(launch_lat, launch_lon, point['lat'], point['lon'])
        if dist < min_distance:
            min_distance = dist
            closest = point
    closest['distance'] = min_distance
    return closest

# Launch site
launch_lat = 28.563197
launch_lon = -80.576820

# Find closest
closest_city = find_closest(cities, launch_lat, launch_lon)
closest_rail = find_closest(railways, launch_lat, launch_lon)
closest_highway = find_closest(highways, launch_lat, launch_lon)

# Add all to map
for point in [closest_city, closest_rail, closest_highway]:
    # Marker
    folium.Marker(
        [point['lat'], point['lon']],
        popup=point['label'],
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(site_map)

    # Line
    folium.PolyLine(
        locations=[[launch_lat, launch_lon], [point['lat'], point['lon']]],
        color='gray',
        dash_array='5,10',
        weight=2
    ).add_to(site_map)

    # Distance label
    folium.Marker(
        [(launch_lat + point['lat'])/2, (launch_lon + point['lon'])/2],
        icon=DivIcon(
            icon_size=(150,36),
            icon_anchor=(0,0),
            html='<div style="font-size: 12px; color:#555;"><b>{:.2f} km to {}</b></div>'.format(point['distance'], point['label']),
        )
    ).add_to(site_map)

# Display map
display(site_map)
```
